```python
# Calculate average of age for each class
df_titanic.groupby('pclass')['age'].mean()
```

```python
# Custom function to return age depending on pclass
def impute_age(cols):
  age = cols[0]
  pclass = cols[1]
  if pd.isnull(age):
    if pclass==1:
      return 38
    elif pclass==2:
      return 29
    else:
      return 25
  else:
    return age
```

```python
df_titanic['age'] = df_titanic[['age','pclass']].apply(impute_age,axis=1)
```
